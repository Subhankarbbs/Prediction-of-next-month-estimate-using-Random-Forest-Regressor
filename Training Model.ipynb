```python
### Data Preprocessing
```

```python
df = pd.read_csv(r'Dataset_Cleaned.csv')
df
```

```python
#Encoding the disttrict
from sklearn.preprocessing import LabelEncoder

le_state = LabelEncoder()
df['state_encoded'] = le_state.fit_transform(df['state'])

le_district = LabelEncoder()
df['district_encoded'] = le_district.fit_transform(df['district'])

X = df[['state_encoded', 'district_encoded', 'year', 'month']]
y = df[['Bio_bio_age_5_17', 'Bio_bio_age_17_']]
```

```python
X
```

```python
### Model Training
```

```python
#TRAIN ON ALL DATA
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X, y)
```

```python
### Model Evaluation
```

```python
model = RandomForestRegressor(n_estimators=100, oob_score=True, n_jobs=-1, random_state=42)

model.fit(X, y)

oob_r2 = model.oob_score_
print(f"Final Model Trained on All Data.")
print(f"OOB Score (Estimated R²): {oob_r2:.4f}")
```

```python
target_year = 2025  
target_month = 3
```

```python
print(f"--- PREDICTING FOR: Year {target_year}, Month {target_month} ---")

# Create the input for the model
unique_locations = df[['state', 'district', 'state_encoded', 'district_encoded']].drop_duplicates()
future_df = unique_locations.copy()
future_df['year'] = target_year
future_df['month'] = target_month

# Predict
X_future = future_df[['state_encoded', 'district_encoded', 'year', 'month']]
predictions = model.predict(X_future)

# Show results
results = future_df[['state', 'district', 'year', 'month']].copy()
results['Predicted_bio5-17'] = predictions[:, 0].astype(int)
results['Predicted_bio17plus'] = predictions[:, 1].astype(int)

print(results.head())
```

```python
district_data = results[results['district'] == 'Nicobar']
```

```python
### Model Export
```

```python
features = ['state_encoded', 'district_encoded', 'year', 'month']

artifacts = {
    'model': model,
    'le_state': le_state,
    'le_district': le_district,
    'features': features
}

joblib.dump(artifacts, 'biometric_model_v1.pkl')
print("✅ Saved to 'biometric_model_v1.pkl'")
```
